I only used the casting_data part, containing 7348 images, on which augmentation has already been applied. 

The original dataset only contained a train part and a test one but I decided to split the train set to create a validation set in order to prevent indirect overfitting. Regarding the total number of images, which is fine but not impressive, I chose a validation split of only 0.15. # 2. Simple custom models <a id='simple'></a>

Before dealing with transfer learning, I'll first build more simple custom models in order to have a baseline.

## 2.1 Naive Feedforward Neural Network <a id='naive_fnn'></a>

I start with a very naive FNN, with only two dense layers (the second one is the output) and a dropout instance. Such an architecture is not adapted to computer vision because it will only use a flattened vector to represent the image. That way, it can hardly understand shapes.

```python
model=keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(units=256, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(units=1, activation="sigmoid"),
])

model.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

We can see that even with a model that simple, the number of parameters is still higher than 50M. That will be a problem during all this work. In fact, I have currently troubles with my laptop and the computation time is even higher than it should normally be. However, you'll see that it seems more usual for some epoch. I won't go into the details but know that those are exceptions that I couldn't replicate. I tried to use Google Collab but, surprisingly, it wasn't faster. Because of that, I won't use all batches at each epoch. I will adapt it to the computational time and the results.

```python
model.fit(train,epochs=10,steps_per_epoch=10,validation_data=valid,validation_steps=len(valid))
```

```python
loss_accuracy, test_accuracy = model.evaluate(test)
print(f"Test accuracy : {np.round(test_accuracy,3)}")
```

On the test set, this naive FNN reached an accuracy of 0.741, which is better than I expected. At first, I was afraid that the model was just predicting 'def' all the time and that the data was too unbalanced but, by doing that, the accuracy on the test set would be only 0.634. However, we can see that the training phase was not stable. The validation accuracy is supposed to improve among epochs but didn't here. We can look at that graphically :

```python
acc = model.history.history['accuracy']
val_acc = model.history.history['val_accuracy']
epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
```

## 2.2 Custom Convolutional Neural Network <a id='custom_cnn'></a>

The second step was to improve this naive FNN by adding a convolutional part at the beginning. I added three classic instances with two max pooling.

```python
cnn=keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.Flatten(),
    keras.layers.Dense(units=256, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(units=1, activation="sigmoid"),
])

cnn.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.summary()
```

Even after adding complexity, the number of parameters did not improve a lot because the vast majority of it are contained in the weight matrix between the flattened vector and the dense layer.

```python
cnn.fit(train,epochs=10,steps_per_epoch=10,validation_data=valid,validation_steps=len(valid))
```

```python
loss_accuracy, test_accuracy = cnn.evaluate(test)
print(f"Test accuracy : {np.round(test_accuracy,3)}")
```

After 10 epochs with 10 steps for each, it reached an accuracy of 0.8. This is a bit disappointing, it is not a lot better than the naive FNN. However, the computation time is still manageable so I decide to start the training again, this time with 20 steps per epoch.

```python
cnn.fit(train,epochs=10,steps_per_epoch=20,validation_data=valid,validation_steps=len(valid))
```

The validation accuracy is not way better than with 10 steps. The problem may not be that but the number of epochs. Indeed, we can see that the model is still improving during the last epochs. I try to add 20 more epochs, with 10 steps each this time.

```python
cnn.fit(train,epochs=20,steps_per_epoch=10,validation_data=valid,validation_steps=len(valid))
```

```python
loss_accuracy, test_accuracy = cnn.evaluate(test)
print(f"Test accuracy : {np.round(test_accuracy,3)}")
```

The result is very impressive, with an accuracy of 0.962 on the test set. By adding more epochs, we might reach an even better result but it is now time to look at transfer learning.

# 3. Transfer learning <a id='transfer'></a>
[to the top](#top)

The principle of transfer learning is to use an already trained model, usually very complex and used for a close task, and adapt it to our particular problem by replacing the last layers and train the new ones on our sets. Here, we'll use three very popular image recognition models trained on ImageNet. These are Xception, VGG16 and ResNet50. I Have chosen to use the simplest versions of VGG and ResNet because of my computation time issues and the fact that our task seems to not be so difficult since we already reached a very good performance.

## 3.1 Xception <a id='xception'></a>

```python
xcept = Xception(input_shape = (256, 256, 3), include_top = False, weights = 'imagenet')
```

```python
for layer in xcept.layers:
    layer.trainable = False


xception=keras.Sequential([
    xcept,
    keras.layers.Flatten(),
    keras.layers.Dense(units=256, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(units=1, activation="sigmoid"),
])

xception.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
```

```python
xception.fit(train,epochs=10,steps_per_epoch=10,validation_data=valid,validation_steps=len(valid))
```

```python
loss_accuracy, test_accuracy = xception.evaluate(test)
print(f"Test accuracy : {np.round(test_accuracy,3)}")
```

After 10 epochs, it reached an accuracy of 0.969 on the test set. This is not much better than the custom CNN. However, the model seems to still be learning during the last epochs so I try to add another one.

```python
xception.fit(train,epochs=1,steps_per_epoch=20,validation_data=valid,validation_steps=len(valid))
```

```python
loss_accuracy, test_accuracy = xception.evaluate(test)
print(f"Test accuracy : {np.round(test_accuracy,3)}")
```

The improvement of 0.004 is not very significant so I choose to not continue in that way but move on to the next model.

# 4. Explainability <a id='explain'></a>
[to the top](#top)

We can now observe explainabality. We'll use the ResNet model that reached a 0.990 accuracy on the test set. Here, it will consist in looking at what parts of images are really useful in determining the class. In order to do that, we'll use LIME on eight examples.

```python
t_dict=test.class_indices
classes=list(t_dict.keys())
plt.figure(figsize=(20, 25))

for i in range(8):
    plt.imshow(test[0][0][i])
    plt.subplot(4, 4, i + 1)
    image=test[0][0][i] 
    plt.imshow(image)
    index=int(test[0][1][i])
    class_name=classes[index]
    plt.title(class_name, color='blue', fontsize=18)
    plt.axis('off')
plt.show()
```

I convert the images tensors to arrays in order to use the LIME explainer.

```python
def get_img_array(images):
    
    arrays = []
    for img in images : 
        
        array = keras.preprocessing.image.img_to_array(img)
        arrays.append(array)
    
    return arrays
```

```python
arrays = get_img_array([test[0][0][i] for i in range(8)])
```

```python
explainer = lime_image.LimeImageExplainer()
```

```python
t_dict=test.class_indices
classes=list(t_dict.keys())
plt.figure(figsize=(20, 25))

for i in range(8): 
    plt.subplot(4,4,i+1)
    explanation = explainer.explain_instance(arrays[i].astype('double'), resnet.predict, top_labels=2, hide_color=0, num_samples=100)
    temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=20, hide_rest=True)
    plt.imshow(mark_boundaries(temp, mask))
    index=int(test[0][1][i])
    class_name=classes[index]
    plt.title(class_name, color='blue', fontsize=18)
    plt.axis('off')

plt.show()
```

Here, the explainability function is only showing the parts of the image that really matter to the classification. The results are not really significant, difficult to interpret. However, we can see that the model understood well the different shapes in the casting product because the borders of the important parts are often following those shapes. It is possible to see that even easily when plotting only the mask. Indeed, we easily recognize the product.

```python
t_dict=test.class_indices
classes=list(t_dict.keys())
plt.figure(figsize=(20, 25))

for i in range(8): 
    plt.subplot(4,4,i+1)
    explanation = explainer.explain_instance(arrays[i].astype('double'), resnet.predict, top_labels=2, hide_color=0, num_samples=100)
    temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=20, hide_rest=False)
    plt.imshow(mask)
    index=int(test[0][1][i])
    class_name=classes[index]
    plt.title(class_name, color='blue', fontsize=18)
    plt.axis('off')

plt.show()
```
